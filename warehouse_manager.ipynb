```python
import json
try:
    with open('magazzino.json', 'r') as file:
        warehouse=json.load(file)
except:        
    warehouse={}
sales_register=[]
class unrecognized_command(Exception):
    pass
cmd=None
while cmd!="chiudi":
   cmd=input("inserisci un comando [elenca-aggiungi-vendita-profitti-aiuto-chiudi]: ")
   if cmd not in ['elenca','aggiungi','vendita','aiuto','chiudi', 'profitti']:
       raise unrecognized_command("Comando non riconosciuto")
       
   if cmd=="aggiungi":
      question="si"
      while question!="no":

        products=input("Inserisci un prodotto in magazzino: ")
        while len(products)==0:
            print("Devi inserire il nome del prodotto")
            products=input("Inserisci il nome del prodotto: ")
        try:
            quantity=int(input("inserisci la quantità di articoli: "))  
        except ValueError:
            print("Devi inserire un valore numerico intero per la quantità")
            continue
        try:
            in_price=float(input("inserisci il prezzo di acquisto: "))
            while in_price<=0:
                print("il prezzo di acquisto deve essere un numero positivo")
                in_price=float(input("inserisci il prezzo di acquisto: "))
        except ValueError:
            print("Devi inserire un valore numerico per il prezzo di acquisto")
            continue
        try:
            out_price=float(input("inserisci il prezzo di vendita: "))
        except ValueError:
            print("Devi inserire un valore numerico per il prezzo di vendita")
            continue
        warehouse[products]={"quantity":quantity,"in_price":in_price,"out_price":out_price}
        print(f"AGGIUNTO: {products} x {quantity} pz")
         
        question=input("vuoi aggiungere un altro articolo? [si/no]")
   
   elif cmd=="elenca":
        for k,v in warehouse.items():
            print(f"PRODOTTO: {k} - PREZZO DI ACQUISTO: {v['in_price']} € - QUANTITA': {v.get('quantity', 0)} pz")
        for i in range(len(sales_register)):
            print(f"REGISTRO VENDITE: PRODOTTO:{sales_register[i]['prodotto'] } - PREZZO DI VENDITA: {sales_register[i]['prezzo_vendita'] } € - QUANTITA': {sales_register[i]['tot_venduto']} - INCASSO: {sales_register[i]['incasso']}€")
            total_collection=sum([item["incasso"] for item in sales_register])
            print(f"L'INCASSO TOTALE E' : {total_collection} €")
   
   elif cmd=="vendita":
       selling_product=input("inserisci il prodotto da vendere: ")
       quantity_sold=int(input("inserisci la quantità del prodotto da vendere: "))
       if selling_product in warehouse:

        if quantity_sold<=warehouse[selling_product]["quantity"]:
            selling_price=warehouse[selling_product]["out_price"]
            tot_sold=quantity_sold
            collection=selling_price*tot_sold
            sales_register.append({"prodotto":selling_product, "prezzo_vendita":selling_price, "tot_venduto":tot_sold, "incasso":collection})
            warehouse[selling_product]["quantity"]-=quantity_sold
            
            print(f"REGISTRO VENDITE: PRODOTTO: {selling_product} - TOTALE VENDUTO: {tot_sold} - INCASSO: {collection} €")
    
        else:
            print("Quantità di prodotto non disponibile in magazzino")
       
       else:   
           print("Il prodotto selezionato non è presente in magazzino")          
  
   elif cmd=="profitti":
       total_collection = sum([item["incasso"] for item in sales_register])
       total_cost=sum([item['tot_venduto'] * float(warehouse.get(item['prodotto'], {}).get('in_price',0)) for item in sales_register])
       profit=total_collection-total_cost
       print(f"PROFITTI : {profit} €")


   elif cmd=="aiuto":
        cmd_1="aggiungi: aggiungi un prodotto al magazzino"  
        cmd_2="elenca: elenca i prodotti in magazzino"
        cmd_3="vendita: registra una vendita effettuata"
        cmd_4="profitti: mostra i profitti totali"
        cmd_5="aiuto: mostra i possibili comandi"
        cmd_6="chiudi: esci dal programma"          
        print(f"LA LISTA COMANDI E' LA SEGUENTE:\n {cmd_1} \n {cmd_2} \n {cmd_3} \n {cmd_4} \n {cmd_5} \n {cmd_6} ")
        break
   
   elif cmd=='chiudi':
       print("Arrivederci")
       break

with open ('magazzino.json', 'w') as file:
    json.dump(warehouse,file)  
       
```
